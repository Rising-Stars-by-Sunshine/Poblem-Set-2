#Part 1

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/naramendia/portfolio/main/data/Queried_Data/JustKoreaGDP.csv'
df = pd.read_csv(url)
df.head(5)
```

```python
df.head()
```

#Part 2

```python
df.columns
```

```python
df.dtypes
```

#Part 3

```python
df['GDPg'] = df["GDP"].pct_change(1)
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
cut = 0
```

```python
df['Positiveg'] = df['GDPg'].apply(lambda x: 1 if x>= cut else 0)
df.head()
```

#Part 4

```python
df['GDPg_past'] =df['GDPg'].shift(1)
df.head()
```

```python
window = 10
```

```python
df['GDPg_past_ma10']=df['GDPg_past'].rolling(window=window,min_periods=1).mean()
df.head(20)
```

#Part 5

```python
from sklearn.model_selection import TimeSeriesSplit
tss = TimeSeriesSplit()
for train_idx, test_idx in tss.split(df):
  pass

train_df = df.filter(items=train_idx,axis=0)
test_df = df.filter(items=test_idx,axis=0)
```

```python
test_df.head()
```
